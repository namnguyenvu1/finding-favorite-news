```python
# URL of the webpage you want to retrieve
url = "https://www.realtor.com/realestateandhomes-search/Austin_TX?view=map"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, 'html.parser')

    # Print the title of the page
    print(soup.title.text)

    # You can also extract other elements from the page as needed
    # For example, to print all the links on the page:
    for link in soup.find_all('a'):
        print(link.get('href'))
else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```

```python
from bs4 import BeautifulSoup

# Load the content from the file
with open('znews.txt', 'r', encoding='utf-8') as file:
    content = file.read()

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(content, 'html.parser')

# Extract titles and intro text
titles = soup.find_all('p', class_='article-title')

# Display the extracted titles and descriptions
for title in titles:
    # Find the link inside the <p> tag
    a_tag = title.find('a')
    if a_tag:
        print("Title:", a_tag.text.strip())
        print("Link:", a_tag['href'])
        print()
```

```python
from googletrans import Translator

translator = Translator()

# Assuming `titles` is a list of titles in Vietnamese
translated_titles = []
for title in titles:
    a_tag = title.find('a')
    if a_tag:
        vietnamese_title = a_tag.text.strip()
        translated = translator.translate(vietnamese_title, src='vi', dest='en')
        translated_titles.append({
            'title': translated.text,
            'link': a_tag['href']
        })

for t in translated_titles:
    print("Translated Title:", t['title'])
    print("Link:", t['link'])
    print()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Extract titles from the list
titles_texts = [t['title'] for t in translated_titles]

# Convert titles to TF-IDF vectors
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(titles_texts)

# Cluster titles into topics
num_clusters = 20  # You can change the number based on your data
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)

# Group titles by their assigned clusters
topics = {}
for i, label in enumerate(kmeans.labels_):
    if label not in topics:
        topics[label] = []
    topics[label].append(translated_titles[i])

# Summarize topics and select top 5 titles for each
for topic_id, topic_titles in topics.items():
    print(f"Topic {topic_id + 1}:")
    for t in topic_titles[:5]:  # Show top 5 titles
        print(f"  Title: {t['title']}")
        print(f"  Link: {t['link']}")
    print()
```

```python
favorite_topics = ["technology", "sports"]  # Define your favorite topics here

# Filter the news based on favorite topics
filtered_news = []
for topic_id, topic_titles in topics.items():
    for t in topic_titles:
        if any(favorite in t['title'].lower() for favorite in favorite_topics):
            filtered_news.append(t)

# Display filtered news
print("News matching your favorite topics:")
for t in filtered_news:
    print(f"Title: {t['title']}")
    print(f"Link: {t['link']}")
    print()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Define your favorite topics
favorite_topics = ["technology", "sports", "politics"]

# Combine all favorite topics into a single string
favorite_topics_combined = " ".join(favorite_topics)

# Create a list of all titles and add the favorite topics as an extra "document"
titles_texts = [t['title'] for t in translated_titles]
documents = titles_texts + [favorite_topics_combined]

# Convert titles and favorite topics to TF-IDF vectors
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)

# Calculate cosine similarity between favorite topics and each title
similarities = cosine_similarity(X[-1], X[:-1]).flatten()

# Sort titles by similarity
sorted_indices = similarities.argsort()[::-1]
filtered_news = [translated_titles[i] for i in sorted_indices if similarities[i] > 0.1]  # You can adjust the threshold

# Display filtered news
print("News matching your favorite topics:")
for t in filtered_news:
    print(f"Title: {t['title']}")
    print(f"Link: {t['link']}")
    print()
```

```python
from transformers import BertTokenizer, BertModel
import torch
from sklearn.metrics.pairwise import cosine_similarity

# Load BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def encode_text(text):
    inputs = tokenizer(text, return_tensors='pt', max_length=512, truncation=True, padding='max_length')
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).numpy()

# Encode favorite topics and titles
favorite_topics_combined = " ".join(favorite_topics)
favorite_embedding = encode_text(favorite_topics_combined)

title_embeddings = []
for t in translated_titles:
    title_embedding = encode_text(t['title'])
    title_embeddings.append(title_embedding)

# Calculate cosine similarity between favorite topics and each title
similarities = [cosine_similarity(favorite_embedding, title_emb)[0][0] for title_emb in title_embeddings]

# Sort titles by similarity
sorted_indices = sorted(range(len(similarities)), key=lambda i: similarities[i], reverse=True)
filtered_news = [translated_titles[i] for i in sorted_indices if similarities[i] > 0.5]  # Adjust threshold as needed

# Display filtered news
print("News matching your favorite topics:")
for t in filtered_news:
    print(f"Title: {t['title']}")
    print(f"Link: {t['link']}")
    print()
```
